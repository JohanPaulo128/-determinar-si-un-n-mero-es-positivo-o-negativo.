## Formula

```python
# Para cuando es verdadero(SI)
# Calcular numeros positivos, negativos con if else
# Datos de entrada NUM
# Mostrar numero negativo si es falso
# Si es verdadero mostrar positivo
NUM=int(input("Ingrese un numero: "))
if NUM<0:
    R="El numero es negativo"
else:
    R="El numero es positivo"
print(R)
print("Fin del programa ")
print("Programa realizado por Johan Paulo 0016 NL 4")
```
